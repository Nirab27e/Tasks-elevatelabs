```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv(r"C:\Users\DELL\Downloads\Titanic-Dataset.csv")
```

```python
print("DATA INFORMATION")
print(df.info())
print("MISSING VALUES",df.isnull().sum())
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())
```

```python
if 'Embarked' in df.columns:
    df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
if 'Cabin' in df.columns:
    df=df.drop(columns=['Cabin'])
```

```python
cate_cols=df.select_dtypes(include='object').columns.tolist()
if 'Name' in cate_cols:
    cate_cols.remove('Name')
if 'Ticket' in cate_cols:
    cate_cols.remove('Ticket')

df=pd.get_dummies(df,columns=cate_cols,drop_first=True)
```

```python
num_cols=df.select_dtypes(include=["int64","float64"]).columns.tolist()
num_cols=[col for col in num_cols if col not in ['PassengerId','Survived']]
scaler=StandardScaler()
df[num_cols]=scaler.fit_transform(df[num_cols])
```

```python
plt.figure(figsize=(12,6))
for i,col in enumerate(num_cols,1):
    plt.subplot(1,len(num_cols),i)
    sns.boxplot(y=df[col])
    plt.title(col)
plt.tight_layout()
plt.show()
    
```

```python
for col in num_cols:
    l=df[col].quantile(0.25)
    u=df[col].quantile(0.75)
    IQR=u-l
    low=l-1.5*IQR
    upp=u+1.5*IQR
    df=df[(df[col]>=low) & (df[col]<=upp)]
print("CLEANED DATASET:",df.shape)
print(df.head())
```
